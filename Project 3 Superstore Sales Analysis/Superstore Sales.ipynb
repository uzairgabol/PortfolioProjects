#### Q1. What is the overall sales trend? 

```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
df_trend = df.groupby('month_year').sum()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend.index,df_trend['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

####  Q2. Which are the Top 10 products by sales?

```python
prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'].sort_values(ascending = False))
```

```python
#Top 10 Products
prod_sales[:10]
```

#### Q3. Which are the Most Selling Products?

```python
prod_quant = pd.DataFrame(df.groupby('product_name').sum()['quantity'].sort_values(ascending = False))
```

```python
# Top 10 Most Selling Product
prod_quant[:10]
```

#### Q4. Which is the most preferred Ship Mode?

```python
ship_mode = pd.DataFrame(df.groupby('ship_mode').count()['order_id'])
```

```python
ship_mode.sort_values(by='order_id',ascending=False)
```

```python
plt.bar(ship_mode.index,ship_mode.order_id)
```

```python
sns.countplot(df['ship_mode'])
```
